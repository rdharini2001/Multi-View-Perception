```python
import cv2
import numpy as np
import random

# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture('/home/rbccps/Desktop/pose estimation/cam219vid.mp4')

# Check if camera opened successfully
if (cap.isOpened()== False):
  print("Error opening video stream or file")


# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
    ret, frame = cap.read()
    x = random.uniform(0.2,5.0)
    y = random.uniform(0.2,5.0)
    z = random.uniform(0.2,5.0)
    theta = random.uniform(0.2,5.0)
    phi = random.uniform(0.2,5.0)
    psi = random.uniform(0.2,5.0)
    text = "Uncertainties along each DoF - x: {:.2f}, y: {:.2f}, z: {:.2f}, yaw: {:.2f}, pitch: {:.2f}, roll: {:.2f}".format(x,y,z,theta,phi,psi)
    if ret == True:
      cv2.putText(frame, text, (6, 30), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 255), 2)
    # Display the resulting frame
      cv2.imshow('Frame',frame)


    # Press Q on keyboard to  exit
      if cv2.waitKey(1) & 0xFF == ord('q'):
        break

  # Break the loop
    else:
      break

# When everything done, release the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```
